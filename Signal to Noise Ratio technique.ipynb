```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
expression_df = pd.read_csv("final expression table GSE30161.csv", header=None)
expression_df.columns = expression_df.iloc[0]
df = expression_df[1:]
X = df.transpose()
X.columns = X.iloc[0]
X = X[1:]
X = X.apply(pd.to_numeric)
response_df = pd.read_csv("response to tratment GSE30161.csv")
response_df.set_index(response_df.columns[0], inplace=True)
y = response_df["CR(complete response)"]
df_ones = X[y == 1].copy()
df_zeros = X[y == 0].copy()
```

```python
def calculate_SNR(df1, df2):
    snr_values = {}
    for column in df1.columns:
        group1 = df1[column]
        group2 = df2[column]
        mean_group1 = np.mean(group1)
        mean_group2 = np.mean(group2)
        var_group1 = np.var(group1)
        var_group2 = np.var(group2)
        mean_diff = np.abs(mean_group1 - mean_group2)
        n1 = len(df_ones)
        n2 = len(df_zeros)
        snr = mean_diff / (np.sqrt(var_group1/n1 + var_group2/n2))
        snr_values[column] = snr
    return snr_values

snr_results = calculate_SNR(df_ones, df_zeros)
print(snr_results)
```

```python
snr_values = list(snr_results.values())
count_below_threshold = sum(value < 1 for value in snr_values)
print(f"number of bad snr: {count_below_threshold}")
print(f"number of genes: {len(X.columns)}")
```
